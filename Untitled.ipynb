```python
train.products_purchased.replace({np.nan: 0}, inplace = True)
```

```python
train['created_at'] = pd.to_datetime(train['created_at'], format='%Y-%m-%d')
```

```python
train['signup_date'] = pd.to_datetime(train['signup_date'], format='%Y-%m-%d')
```

```python
train['diff_create_sign'] = (train['created_at']-train['signup_date']).dt.days
```

```python
train.diff_create_sign.replace({np.nan: -1}, inplace = True)
```

```python
train['avg_activity_var'] = (train['user_activity_var_1'] + train['user_activity_var_2'] + train['user_activity_var_3'] + train['user_activity_var_4'] + train['user_activity_var_5'] + train['user_activity_var_6'] + train['user_activity_var_7'] + train['user_activity_var_8'] + train['user_activity_var_9'] + train['user_activity_var_10'] + train['user_activity_var_11'])/11
```

```python
train['campaign_to_purchase_ratio'] = train['products_purchased']/(train['campaign_var_1'] + train['campaign_var_2'])
```

```python
train.drop('signup_date',axis=1,inplace=True)
train.drop('created_at',axis=1,inplace=True)
```
